## **Training Part**

```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size= 0.2, random_state=0)
```

```
X_train.shape
```

```
X_test.shape
```

```
Y_train.shape
```

```
Y_test.shape
```

```
import tensorflow as tf
```

```
import tensorflow as tf
model = tf.keras.models.Sequential([
                                   # First Hidden Layer
                                   tf.keras.layers.Dense(units=16,input_dim = 20, activation='relu'),
                                   tf.keras.layers.Dense(units= 18, activation='relu'),
                                   
                                   
                                   # Output Layer 
                                   tf.keras.layers.Dense(units=1)

])
```

```
model.summary()
```

```
model.compile(
    optimizer= 'adam',
    loss = 'mean_squared_error'
)
```

```
history = model.fit(X_train, Y_train , steps_per_epoch=10, epochs=300, verbose=1)
```

```
model.save('/content/drive/My Drive/Machine learning')
```
